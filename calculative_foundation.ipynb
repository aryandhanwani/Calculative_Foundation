```python
#Part A: Vector & Matrix Fundamentals

#1. Represent each student's subject scores as a vector.
vectors = np.array(df)

#2.1 Norm-1 and Norm-2 of vectors.
A = vectors[0]
B = vectors[1]

print("Vector A:", A)
print("Vector B:", B,"\n")

#Norm 1
norm1_A = np.linalg.norm(A, ord=1)
norm1_B = np.linalg.norm(B, ord=1)

print("L1 Norm of A:", norm1_A)
print("L1 Norm of B:", norm1_B,"\n")

#Norm 2
norm2_A = np.linalg.norm(A, ord=2)
norm2_B = np.linalg.norm(B, ord=2)

print("L2 Norm of A:", norm2_A)
print("L2 Norm of B:", norm2_B,"\n")

#2.2 Dot product and angle between two students' score vectors.
dot_product = np.dot(A, B)
print("Dot Product:", dot_product,"\n")

#2.3 Cross product (for 3D selected subjects).
A_3D = A[:3]
B_3D = B[:3]

cross_product = np.cross(A_3D, B_3D)
print("Cross Product:", cross_product,"\n")

#3 Find the projection of one vector onto another.
projection = (dot_product / (norm2_B**2)) * B
print("Projection of A onto B:", projection)
```

```python
#Part B: Matrix Operations

#4.1 Matrix addition and multiplication.

M = np.array(df)
print("Matrix Shape:", M.shape)
print(M[:5],"\n")

#addition
M_add = M + M
print("Matrix Addition:\n", M_add[:5],"\n")

#Multiplication
M_T = M.T
M_mul = np.dot(M, M_T)
print("Matrix Multiplication Shape:", M_mul.shape)

#4.2 Transpose
transpose = M.T
print("Transpose Shape:", transpose.shape,"\n")

#4.3 Determinant
square_matrix = M[:4, :4]
det = np.linalg.det(square_matrix)
print("Square Matrix:\n", square_matrix)
print("Determinant:", det)


```

```python
#Part D: Eigenvalues & Eigenvectors

#7. Compute the eigenvalues and eigenvectors of the covariance matrix
eigenvalues, eigenvectors = np.linalg.eig(square_matrix)

print("Eigenvalues:\n", eigenvalues)
print("\nEigenvectors:\n", eigenvectors,"\n")

cov_matrix = np.cov(M, rowvar=False)
print("Covariance Matrix:\n", cov_matrix,"\n")

#8. Perform LU Decomposition of the dataset matrix.
from scipy.linalg import lu
P, L, U = lu(square_matrix)
print("L:\n", L)
print("\nU:\n", U)

```

```python
#10. Apply Principal Component Analysis (PCA) to reduce the dataset from multiple subjects to 2 dimensions.

X = df[["Math", "Science", "English", "Computer"]].values

mean = np.mean(X, axis=0)
X_centered = X - mean
cov_matrix = np.cov(X_centered, rowvar=False)
print("Covariance Matrix:\n", cov_matrix,"\n")

eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
print("Eigenvalues:\n", eigenvalues)
print("\nEigenvectors:\n", eigenvectors)


idx = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[idx]
eigenvectors = eigenvectors[:, idx]

W = eigenvectors[:, :2]
X_pca = np.dot(X_centered, W)
print("PCA Result (2D):\n", X_pca[:5])
```

```python
# 11 Apply Linear Discriminant Analysis (LDA) to classify students into "Above Average" and "Below Average" categories.

subjects = ["Math", "Science", "English", "Computer"]
df["Average"] = df[subjects].mean(axis=1)
overall_avg = df["Average"].mean()
df["Category"] = df["Average"].apply(lambda x: "Above Average" if x >= overall_avg else "Below Average")
print(df[["Average", "Category"]].head())
```
